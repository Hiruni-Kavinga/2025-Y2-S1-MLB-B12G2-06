```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset from the raw data folder
df = pd.read_csv('healthcare-dataset-stroke-data.csv')

# Display the first few rows to verify loading
print(df.head())
```

```python
# Check for missing values in the dataset
print("Missing values per column:")
print(df.isnull().sum())
```

```python
# Specifically for BMI
missing_before = df['bmi'].isnull().sum()
print(f"\nMissing values in BMI before imputation: {missing_before}")
```

```python
# EDA Visualization: Histogram of BMI before imputation (skips NaNs automatically)
plt.figure(figsize=(10, 5))
sns.histplot(df['bmi'], kde=True, color='blue')
plt.title('BMI Distribution Before Imputation')
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Impute missing BMI values with the mean (simple and effective for this numerical feature)
mean_bmi = df['bmi'].mean()
df['bmi'].fillna(mean_bmi, inplace=True)

# Verify after imputation
missing_after = df['bmi'].isnull().sum()
print(f"Missing values in BMI after imputation: {missing_after}")
print(f"Mean BMI used for imputation: {mean_bmi:.2f}")
```

```python
# EDA Visualization: Histogram of BMI after imputation
plt.figure(figsize=(10, 5))
sns.histplot(df['bmi'], kde=True, color='green')
plt.title('BMI Distribution After Imputation')
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Save processed dataset
df.to_excel("healthcare-stroke-data-cleaned.xlsx", index=False)
print("Processed dataset saved for pipeline integration.")
```
